```python
class BankAccount():
    def __init__(self, bank_name, name, balance):
        self.bank_name = bank_name
        self.name = name
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount

    def check_balance(self):
        print(f'Your current balance is {self.__balance}')

    def details(self):
        print(f'Mr.{self.name}, Your balance in {self.bank_name} bank account is {self.__balance} ')

account = BankAccount('Meezan', 'Haseeb', 0)
account.deposit(1000)
```

```python
account.check_balance()
```

```python
account.bank_name = "HBL"
```

```python
account.bank_name
```

```python
account.withdraw(500)
```

```python
account.check_balance()
```

```python
account.details()
```

```python
account.__balance = 123
```

```python
account.check_balance()
```

```python
class Student():
    def __init__(self, name):
        self.name = name
        self.__marks = 0

    def get_marks(self):
        print(f'{self.name} your marks is {self.__marks}')

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks

student = Student('haseeb')
```

```python
student.set_marks(89)
```

```python
student.get_marks()
```

```python
class ShoppingCart():
    def __init__(self, total_price):
        self.__total_price = 0
        self.items = []

    def add_item(self, items, price):
        self.items.append(items)
        self.__total_price += price

    def get_total(self):
        return self.__total_price
```

```python
a = ShoppingCart(0)
```

```python
a.add_item('bread', 150)
```

```python
a.get_total()
```

```python
a.add_item('eggs', 150)
```

```python
a.get_total()
```

```python
def count_by(x, n):
    result = []
    for i in range(1, n+1):
        result.append(x * i)

    return result
```

```python
count_by(3, 6)
```

```python
def double_char(s):
    result = ""
    for char in s:
        result += char*2 
        
    return result
```

```python
double_char('hello')
```

```python
n = ['my', 'name', 'haseeb']
def smash(words):
    result = ""
    result += " ".join(words)
    
    return result
```
