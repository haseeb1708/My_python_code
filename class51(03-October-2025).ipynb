```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Supermart Grocery Sales.csv', encoding = 'utf-8')
```

```python
df
```

Q1. Which region has the most sales?

```python
most_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(1)
most_sales
```

Answer:
'West' region has the most sales. '4798743'

Q2. Which region has the highest profit?

```python
highest_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False).head(1)
highest_profit
```

Answer: 'West' region has the highest profit. '1192004.61'

Q3. Who is the most frequent customer?

```python
most_frequent_customer = df['Customer Name'].value_counts().sort_values(ascending=False).head(1)
most_frequent_customer
```

Answer: Amrish is the most frequent customer with '227' orders

Q4. Among the categories."Beverages", 'Bakery' and 'Snacks'.Which is most profitable in south region?

```python
south_data = df[df['Region'] == 'South']
south_data
```

```python
filtered_data = south_data[south_data['Category'].isin(['Beverages', 'Bakery', 'Snacks'])]
filtered_data
```

```python
most_profitable = filtered_data.groupby('Category')['Profit'].sum().sort_values(ascending=False).head(1)
most_profitable
```

Answer: 'Snacks' are most profitable in south region with '94979.87' profit.

Q5. Plot the sales of the store according to order date. Using linear regression, find if sales is increasing or decreasing?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=False, infer_datetime_format=True)
df['Order Date']
```

```python
sales_by_date = df.groupby('Order Date')['Sales'].sum().reset_index()
sales_by_date
```
