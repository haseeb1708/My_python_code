```python
def separate_positive_negative(numbers):
    positive = []
    negative = []
    
    for num in numbers:
        if num >= 0:
            positive.append(num)
        else:
            negative.append(num)
    
    print("Positive Numbers:", positive)
    print("Negative Numbers:", negative)

numbers = [-10, 3, -2, 7, -8, 5, 0, -6, 4]
separate_positive_negative(numbers)
```

```python
class Employee():
    def __init__(self, name, salary):
        self._name = name
        self._salary = salary

    def display(self):
        print(f'Employee: {self._name}\n salary: {self._salary}')

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus

    def update_salary(self):
        self._salary += self.bonus
```

```python
haseeb = Manager('Haseeb', 30000, 10000)
```

```python
haseeb.display()
```

```python
class Student():
    def __init__(self):
        self.__grades = []

    def add_grade(self, grade):
        if 0<= grade <=100:
            self.__grades.append(grade)
        else:
            print('Invalid Value')

    def get_average(self):
        if self.__grades:
            return sum(self.__grades) / len(self.__grades)
        else:
            return 0
```
