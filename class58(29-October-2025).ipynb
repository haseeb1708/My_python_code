```python
import matplotlib.pyplot as plt

# Data for the bar chart
categories = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
values = [320, 450, 500, 480, 600]
colors = ['skyblue', 'lightgreen', 'salmon', 'gold', 'orchid']
```

```python
fig = 'colored_bar_chart.png'
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales (s)')
plt.bar(categories, values, color=colors)
plt.savefig(fig) 
print(f'Colored bar chart saved as {fig}')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Ali', 'Rahim', 'Hassam', 'Faizan']
values1 = [10, 15, 7, 20]
values2 = [12, 18, 6, 15]
values3 = [8, 10, 12, 9]
```

```python
x = np.arange(len(categories))
x
```

```python

x = np.arange(len(categories))
width = 0.2

fig, ax = plt.subplots()

bars1 = ax.bar(x - width, values1, width, label='Group 1')
bars2 = ax.bar(x , values2, width, label='Group 2')
bars2 = ax.bar(x + width, values3, width, label='Group 2')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Grouped Bar Plot')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.legend()

plt.show()
```
