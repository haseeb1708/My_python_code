# Class and Instance Attributes in Python


```python
class Person:
    # class attribute
    brain_count = 1
    
    def __init__(self, name, age, country):

        #Instance Attributes
        self.name = name
        self.age = age
        self.country = country

    def display_info(self):
        print(f'My name is {self.name}. I am {self.age} years old. I live in {self.country}')
```

```python
class Person:
    # class attribute
    brain_count = 1
    
    def __init__(self, name, age, country):

        #Instance Attributes
        self.name = name
        self.age = age
        self.country = country

        
        
    @classmethod
    def class_method(cls, name, age, country):
        return cls(name,age, country)
    # Factory Method
```

```python
kashan = Person('Kashan', 20, 'Pakistan')
```

```python
#Factory Method
```

# Static Method

```python
class Person:
    # class attribute
    brain_count = 1
    
    def __init__(self, name, age, country):

        #Instance Attributes
        self.name = name
        self.age = age
        self.country = country
    
    @staticmethod
    def static_method(text):
        print(text.upper())
    #utitlity method
        
    @classmethod
    def class_method(cls, name, age, country):
        return cls(name,age, country)
    # Factory Method / Alternative constructor
```

```python
Person.static_method('ali khan')
```
