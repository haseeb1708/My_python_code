# Concatenating Arrays

```python
import numpy as np
arr2d_a = np.arange(25).reshape(5,5)
arr2d_b = np.arange(25,50).reshape(5,5)
```

```python
arr2d_a
```

```python
arr2d_b
```

```python
vertical_joined_array = np.concatenate((arr2d_a,arr2d_b),axis=0)
vertical_joined_array
```

```python
horizontal_joined_array = np.concatenate((arr2d_a,arr2d_b),axis=1)
horizontal_joined_array
```

```python
arr1 = np.arange(20).reshape(4,5)
arr2 = np.arange(20,40).reshape(5,4)
print(arr1)
print(arr2)
```

```python
np.concatenate((arr1,arr2), axis=1)
# Shape / dimension should be same for concatenation
```
