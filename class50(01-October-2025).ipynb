```python
import time
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
# Initialize geolocator
geolocator = Nominatim(user_agent="state_finder")

# To avoid being blocked, use a rate limiter (1 request per second is safe)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

def get_state(city, country):
    try:
        location = geocode(f"{city}, {country}")
        raise f'alikhan...................{location}'
        if location and 'state' in location.raw['address']:
            return location.raw['address']['state']
        elif location and 'state_district' in location.raw['address']:
            return location.raw['address']['state_district']
        else:
            return None
    except Exception as e:
        print(f"Error for {city}, {country}: {e}")
        return None
```

```python
df.apply(lambda row: get_state(row['CITY'], row['COUNTRY']), axis=1)
```
