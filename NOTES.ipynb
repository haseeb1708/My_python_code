### SET

set unordered hota hai , **mutable** hota hai , lekin duplicate items allow nahi karta.

**Set Methods:**
- add()
- remove()
- discard()
- union()
- intersection()
- difference()
- clear()

```python
my_set = {1, 2, 3, 4}
```

```python
my_set.add(5)
my_set
```

```python
my_set.remove(5)
my_set
```

```python
my_set.discard(4)
my_set
```

```python
my_set.union({4, 5})
```

```python
a = {1, 2, 3, 4, 5, 6, 7, 8}
my_set.intersection(a)
my_set
```

```python
my_set.difference()
```

```python
my_set.clear()
```

```python
my_set
```

### DICTIONARY

Dictionary key-value pairs ka collection hai. Ye **unordered** (Python 3.6 se ordered) aur **mutable** hota hai.

**Dictionary Methods:**
- get()
- keys()
- values()
- items()
- update()
- pop()
- clear()

```python
my_dict = {"name" : "Haseeb" , "age" : "17" , "course" : "Python"}
```

```python
my_dict.get("name")
```

```python
my_dict.keys()
```

```python
my_dict.values()
```

```python
my_dict.items()
```

```python
my_dict.update({"institute" : "SMIT"})
```

```python
my_dict
```

```python
my_dict.pop("age")
```

```python
my_dict.popitem()
```

```python
my_dict
```

```python
my_dict.clear()
```

### LOOPS

**FOR LOOP:**
- kisi sequence(jaise list, range, string)ke elements ko ek ek karke access karta hai

```python
#for example:
for i in range(5):
    print(i)
```

**WHILE LOOP:**
- jab tak condition True hoti hai, loop chalta hai.

```python
#for example
i = 0 
while i > 5:
    print(i)
    i += 1
```

**BREAK MOMENTS:**
- loop ko beech mai hi rok deta hai.
