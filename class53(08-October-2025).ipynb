### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels

```python
# s1['Anday': 'Ali']
# s1.iloc[inclusive:exclusive]
```

```python
# single row
movies.iloc[5]
```

```python
# multiple row
# movies.iloc[starting row : ending row(exclusive)]
movies.iloc[3:10]
movies.iloc[3:10][['title_x']]
```

```python
# fancy indexing
movies.iloc[[0,2,5]]
```

### Selecting both rows and cols

```python
# dataframe.iloc[row,col]>>>> dataframe.iloc[s_row:e_row(exclusive),s_col:e_col(exclusive)]
movies.iloc[5:10 ,2:5 ]
```

```python
# dataframe.loc[row,col]>>>> dataframe.loc[s_row:e_row(inclusive),s_col:e_col(inclusive)]

movies.loc[4: 9, 'story': 'actors']
# movies.columns
```
